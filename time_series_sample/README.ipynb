## 異常検知手法の例についての概要と対応するサンプルコード

(参考: https://engineer-fumi.hatenablog.com/entry/2018/09/18/161914)

異常検知の種類：
- 外れ値検知：
    - ホテリング法 (`./hotelling.ipynb`)・・・・平均からどれだけずれているかを見る．正規分布が前提．
    - k近傍法 (`./K Nearest Neighbor.ipynb`)・・・・距離を見る．密度が違うとなかなか距離が分からなくなる．
    - LOF (`./Local Outlier Factor.ipynb`)・・・・密度を見る．
    - One-Class SVM (`./oneclass_svm.ipynb`)・・・カーネルトリックを使って高次元空間に射影し，異常なものは原点近くに集まる．非線形性も扱える．ただし，パラメータのチューニングが必要．
- 変化点予知：
    - 累積和法 (`./累積和法.ipynb`)・・・・平均から超えたところを累積して，閾値から超えたら異常とする
    - Change Finder (`./change_finder.ipynb`)・・・・3つのステップで変化検知を行う．外れ値検知を利用して変化点を見つける方法．外れ値スコアを見ている．オンラインでの変化検知ができる．（changefinderというモジュールがある）
- 異常部位検知：
    - Matrix Profile (`./matrix_profile.ipynb`)・・・・部分時系列データ同士の類似結合(https://qiita.com/k-kotera/items/91c4181d6b4f1b7f434b , https://www.cs.ucr.edu/~eamonn/MatrixProfile.html)
- NNを使った異常検知：
    - AutoEncoder (`./220122A_企業協働_サンプルコード.ipynb`)・・・・EncoderとDecoderの2プロセスで行っている．学習してモデルにデータを入力し，同様に再現できれば正常と判定．計算量が多いので，オンラインには厳しいかも．GUNやCNNを利用できる．
    - AutoEncoder_LSTM (`./AutoEncoder_LSTM.ipynb`)・・・・AutoEncoderに時間依存性を考慮するためのLSTMブロックを追加したサンプルプログラム.

異常検知データの可視化／特徴量抽出：
- `./tsfresh_sample.ipynb`, `./tsfresh_ResultVisualizing.ipynb`・・・・tsfresh を用いた特徴量抽出と可視化のサンプルコード．自動的に多数の時系列特性，統計量を計算するライブラリ：tsfresh を用いて，対象データの統計量を計算することで，回帰または分類タスクの特性の説明力と重要性を評価する

異常検知の実践TIPS：
- データの状態を考慮する・・・・状態を考慮しないと違う原因で異常となってしまう（ON/OFF時やアイドリング時など）
- データのクレンジング・・・・季節性の除去，トレンドの除去，分布の変換
- 判定結果のフィードバックを考慮する・・・・異常判定→フィードバック→検知方法に反映（オンラインで学習するときに必要）
